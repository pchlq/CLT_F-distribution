# Центральная предельная теорема своими руками

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import math
```

# Распределение Фишера 
Общая информация в [Википедии](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%A4%D0%B8%D1%88%D0%B5%D1%80%D0%B0/ "Фишер").

```python
# посчитаем среднее и дисперсию
d1 = 6.
d2 = 5. # минимальное значение степени свободы для расчета дисперсии (d2>4)

# Расчет матожидания (среднего)
E = d2 / (d2-2)

# Расчет дисперсии
D = (2 * d2**2 * (d1 + d2 - 2)) / (d1 * (d2 -2)**2 * (d2 - 4))

print 'E[x] = ', round(E, 2)
print 'D[x] = ', round(D, 2)
```

```python
# установим первые значения аргументов и зададим функцию
dfn, dfd = 6, 5
fish = sts.f(dfn, dfd)

# генерация случайной выборки объемом 1000
r = fish.rvs(1000)

# определяем массив точек для построения плотности распределения
x = np.linspace(sts.f.ppf(0.01, dfn, dfd), sts.f.ppf(0.99, dfn, dfd), 1000)

# вычисляем значение плотности распределения от выборки
pdf = fish.pdf(x)

# строим гистограмму и теоретическую функцию плотности распределения
plt.hist(r, bins=50, normed=True, alpha=0.8)
plt.plot(x, pdf, 'r-', lw=3, label='theoretical PDF')
plt.ylabel('$f(x)$/number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
```

# Оценка распределение выборочного среднего случайной величины при разных объёмах выборок

# n = 5

```python
n=5
sample5=np.array([])# обозначим массив выборочных средних
i=1
while i <= 1000:# цикл генерации случайных выборок количества n из функции
    sample5i = fish.rvs(n) # генерация выборки
    mean5i=sum(sample5i)/n # определение выборочного среднего (в.с.)
    sample5=np.append(sample5,mean5i)# добавляем в массив новое значение в.с.
    i+=1
#определим параметры нормального распределения описывающие массив 1000 в.с. при n=5
E=1.67 # среднее Фишера
D=8.33 # дисперсия Фишера
D5=D/n # дисперсия для нормального распределения приближающего выборку средних
sigma5 = math.sqrt(D5) # сигма для нормального распределения выборки средних
norm5 = sts.norm(E, sigma5) # определяем нормальное распределение с расчетными характеристиками Фишера
pdf5 = norm5.pdf(x[:500]) # получаем плотность распределения
#строим гистограмму массива средних при n=5 и плотность описывающую этот массив нормального распределения
plt.plot(x[:500], pdf5, label='PDF(n=5)')
plt.hist(sample5, bins =80, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
```
